# Sets

* Sets charateristics
    * Sets are unordered
    * Set elements are unique and duplicated elements are not allowed
    * A set itself may be modified, but the elements contained in the set must be of an immutable type

1. Immutable types:
    1. int
    1. float
    1. strings
    1. decimal
    1. bool
    1. tuples
    1. range
2. mutable types:
    2. user defined types
    2. Dictionaries
    2. set
    2. list

Defining a set (two ways)

```python
# way 1
#x = set(<iter>)
x = set(['foo', 'bar', 'baz', 'foo', 'qux']) # either a list of a tuple
# way 2 using {}
x = {'foo', 'bar', 'baz', 'foo', 'qux', 'foo'}
x
```

WARNING!
define an empty set using the set function otherwise an empty set defined with curly braces will be registered as an empty dictionary

# Modyfing a set

#### x1.update(x2[,x3 ...])
add any elements found in x2 that x1 does not already have

```python
>>> x1 = {'foo', 'bar', 'baz'}
>>> x2 = {'foo', 'baz', 'qux'}

>>> x1 |= x2
>>> print(x1)
>>> x1.update(['corge', 'garply'])
>>> x1
```

#### x1.intersection_update(x2[, x3 ...])
retains only elements found in both x1/x2

# >>> x1 = {'foo', 'bar', 'baz'}
>>> x2 = {'foo', 'baz', 'qux'}

>>> x1 &= x2
>>> print(x1)
>>> x1.intersection_update(['baz', 'qux'])
>>> x1

#### x1.difference_update(x2[, x3 ...])
remove elements in x1 that are found in x2

```python
>>> x1 = {'foo', 'bar', 'baz'}
>>> x2 = {'foo', 'baz', 'qux'}

>>> x1 -= x2
>>> print(x1)
>>> x1.difference_update(['foo', 'bar', 'qux'])
>>> x1
```

#### x1.symmetric_difference_update(x2)
reatin elements found in either x1 or x2 but not both

```python
>>> x1 = {'foo', 'bar', 'baz'}
>>> x2 = {'foo', 'baz', 'qux'}
>>> 
>>> x1 ^= x2
>>> print(x1)
>>> 
>>> x1.symmetric_difference_update(['qux', 'corge'])
>>> x1
```
