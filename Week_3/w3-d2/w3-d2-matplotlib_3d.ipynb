```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
```

```python
plt.style.use('dark_background')
```

```python
# create an empty canvas
ax = plt.axes(projection='3d')
```

We will demonstrate 3D plots on a multivariate normal distribution. The first step is to generate a domain for the distribution:

```python
# divide the inetrval <-10,10> into 500 parts
#https://numpy.org/doc/stable/reference/generated/numpy.linspace.html?highlight=linspace#numpy.linspace
x = np.linspace(-10,10,500)
y = np.linspace(-10,10,500)

# a Cartesian product
X,Y = np.meshgrid(x,y)
```

```python
# the parameters of the distribution
# x
mu_x = 0
variance_x = 10

# y
mu_y = 0
variance_y = 10

# a means list
means = [mu_x, mu_y]

# a covariance
cov = [[variance_x, 0], [0, variance_y]]

# a multivariate normal distribution
mn = multivariate_normal(means, cov)
```

```python
# a position matrix
position = np.array([X,Y]).T

# z-coordinates from a probability-density function
Z = mn.pdf(position)
```

### Wireframe

## Warning! 
We need to run following snippets of code in one cell of jupyter notebook for each graph. Otherwise, it won't be rendered properly.

```python
#  create the figure and axes
fig,ax = plt.subplots(figsize=(12.8,9.6))

# set a 3D projection
ax = plt.axes(projection='3d')

# a wireframe
ax.plot_wireframe(X, Y, Z, color='r')
plt.show()
```

### Contour

```python
#  create the figure and axes
fig,ax = plt.subplots(figsize=(12.8,9.6))

# set a 3D projection
ax = plt.axes(projection='3d')

# the surface
ax.plot_surface(X, Y, Z, cmap='jet')
plt.show()
```

```python
#  create the figure and axes
fig,ax = plt.subplots(figsize=(12.8,9.6))

# set a 3D projection
ax = plt.axes(projection='3d')

# the surface
ax.contour(X, Y, Z, 40, cmap='jet') #he number 40 in the contour function below is the number of contour lines to be drawn
plt.show()
```
