```python
# import seaborn
import seaborn as sns

# import matplotlib
import matplotlib.pyplot as plt

# load the built-in tips dataset
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

Isn't based on the "r" valuue?

```python
# a regression line with a confidence inetrval plot
sns.set_style("dark")
sns.lmplot(x='total_bill', y='tip', data = tips)
plt.show()
```

Now if we want to have two regression lines in one chart – one for female tips and one for male tips – we can modify the code above and use the hue parameter.

```python
sns.set_style("dark")
sns.lmplot(x='total_bill',y='tip', data = tips, hue = 'sex')
plt.show()
```

```python
sns.set_style("dark")
sns.lmplot(x='total_bill',y='tip', data = tips, col = 'smoker', hue='sex')
plt.show()
```

## Residual Plot

Another plot that is related to the regression line is a residual plot. We don't need to compute residuals manually and then plot the results. Instead, we can use a built-in function:

```python
sns.set_style("dark")
sns.residplot(x='total_bill',y='tip', data = tips)
plt.show()
```

```python
#strip plot
sns.set_style("dark")
sns.stripplot(x='day',y ='tip',data = tips, size =5, jitter = True)

# overwrite the original y-label from 'tip' to 'tip ($)'
plt.ylabel('tip ($)')
plt.show()
```

```python
#swarm plot
# you can adjust the size of the plot points using size
sns.set_style("dark")
sns.swarmplot(x='day',y='tip',data = tips, hue = 'sex',size=3)

# overwrite the original y-label from 'tip' to 'tip ($)'
plt.ylabel('tip ($)')
plt.show()
```

```python
sns.set_style("dark")
# create a grid with 1 row 2 columns
plt.subplot(1,2,1)

# the first plot
sns.boxplot(x='day',y='tip', data = tips)
plt.ylabel('tip ($)')

plt.subplot(1,2,2)

# the second plot
sns.violinplot(x='day',y='tip', data = tips)
plt.ylabel('tip ($)')
plt.tight_layout()
plt.show()
```

```python
# joint plot with histograms
sns.set_style("dark")
sns.jointplot(x='total_bill', y='tip', data = tips)
plt.show()
```

```python
# joint plot with density estimates
sns.set_style("dark")
sns.jointplot(x='total_bill', y='tip', data = tips, kind = 'kde')
plt.show()
```

```python
#pairplot (what a mess tho)
sns.set_style("dark")
sns.pairplot(tips,hue = 'sex')
plt.show()
```

```python
#heatmap of all the correlations
# compute correlations between features# compute correlations between features
df_corr = tips.corr()

# plot the correlations
sns.set_style("dark")
sns.heatmap(df_corr)
plt.title('Correlation plot')
plt.show()
```
